```python
#Part 1.1

import sqlite3
import pandas as pd

db_path = 'data/chinook.db'
conn = sqlite3.connect(db_path)
customers_df = pd.read_sql_query("SELECT * FROM customers", conn)
customers_df.head(10)
```

```python
#Part 1.2
iris_df = pd.read_json('data/iris.json')
print("Shape of the dataset: ", iris_df.shape)
print("Column names: ", iris_df.columns)
```

```python
#Part 1.3

titanic_df = pd.read_excel('data/titanic.xlsx')
titanic_df.head(5)
```

```python
#Part 1.4
flights_df = pd.read_parquet('flights.parquet')

flights_df.info()
print(flights_df.head(2))
```

```python
#Part 1.5
movie_df = pd.read_csv('data/movie.csv')
movie_df.sample(10)
```

```python
#Part 2.1
iris_df.columns = iris_df.columns.str.lower()
sepal_df = iris_df[['sepallength', 'sepalwidth']]
```

```python
#Part 2.2
above_30_df = titanic_df[titanic_df['Age'] > 30]

gender_counts = titanic_df['Sex'].value_counts()
print(gender_counts)
```

```python
#Part 2.3
flights_subset = flights_df[['Origin', 'Dest', 'Reporting_Airline']]
print(flights_subset.head())
unique_destinations = flights_df['Dest'].nunique()
print("Number of unique destination: ", unique_destinations)

```

```python
#Part 2.4
long_movies_df = movie_df[movie_df['duration'] > 120]
sorted_long_movies = long_movies_df.sort_values(by='director_facebook_likes', ascending=False)
print(sorted_long_movies)
```

```python
#Part 3.1
sepal_df.head(5)
iris_stats = sepal_df.agg(['mean', 'median', 'std'])
print(iris_stats)
```

```python
#Part 3.2
agg_stats = titanic_df['Age'].agg(['min', 'max', 'sum'])
print(agg_stats)
```

```python
#Part 3.3
top_directors = movie_df.groupby('director_name')['director_facebook_likes'].sum().idxmax()
print("Director with most facebook likes: ", top_directors)

longest_movies = movie_df.nlargest(5, 'duration')[['movie_title','director_name', 'duration']]
print(longest_movies)
```
